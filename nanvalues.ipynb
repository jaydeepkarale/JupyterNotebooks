```python
# Handling Mising Values In A Datasett
import pandas as pd
import numpy as np

df = pd.read_csv('F:/Python/footballml/Missing_Values_Players.csv')
df.head()
```

```python
# Replace all missing values with NaN & make the changes in place
df.replace('?', np.nan, inplace=True)
df
```

```python
# we can now check the total nan values in each column
df.isna().sum()
```

```python
# fill the nan values with a mean value of other columns
# df['Goals Scored'] = df['Goals Scored'].fillna(df['Goals Scored'].mean())
df['Goals Scored'] = pd.to_numeric(df['Goals Scored'], downcast="integer")
df['Goals Scored'] = df['Goals Scored'].fillna(df['Goals Scored'].mean())
df
```
